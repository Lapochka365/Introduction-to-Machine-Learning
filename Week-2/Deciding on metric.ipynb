# Выбор метрики

###### Импорт необходимых библиотек

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.datasets import load_boston
from sklearn.preprocessing import scale
```

###### Загрузка датасета и разбиение на целевую переменную $y$ и признаки $X$

```python
df = load_boston()
X = df.data
y = df.target
```

###### Использование метода k ближайших соседей с метрикой Минковского и кросс-валидацией

```python
# масштабирование признаков
scaled_X = scale(X)
```

```python
# стратегия кросс-валидации
cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# Оценки кросс-валидации при различных p метрики
p_values = np.linspace(1, 10, 200)
scores = np.empty(0)
for p in p_values:
    score = cross_val_score(KNeighborsRegressor(n_neighbors=5, weights='distance', 
                                               metric='minkowski', metric_params={'p': p}), 
                            X, y, scoring='neg_mean_squared_error', cv=cv).mean()
    scores = np.append(scores, score)
```

```python
optim_p = int(p_values[np.where(scores == np.amax(scores))[0]][0])
print('Оптимальное значение p метрики Минковского -', optim_p)
print('Оптимальное качество -', round(-1 * np.amax(scores), 1))
```
